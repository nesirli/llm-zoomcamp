# RAG with `minsearch` — Refactored

The pipeline from `01-rag-starter` has been extracted into two reusable modules:
- `ingest.py` — fetches FAQ documents and builds the `minsearch` index
- `rag_helper.py` — defines `RAGBase`, a class that wires search + LLM into a `.rag()` method

This notebook shows the clean, importable version with no boilerplate.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from zoomcamp.ingest import load_faq_data, build_index
from zoomcamp.rag_helper import RAGBase
```

```python
documents = load_faq_data()
index = build_index(documents)
```

```python
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

openai_client = OpenAI(api_key=api_key)
```

```python
assistant = RAGBase(
    index=index,
    llm_client=openai_client
)
```

```python
answer = assistant.rag("I just discovered the course. Can I join now?")
print(answer)
```
